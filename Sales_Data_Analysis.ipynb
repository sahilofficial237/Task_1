```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

print("üìÇ Upload your CSV file:")
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

print("\n‚úÖ Dataset Loaded Successfully!\n")
print("üîπ First 5 Rows:")
display(df.head())

print("\nüîπ Dataset Info:")
print(df.info())

print("\nüîπ Missing Values:")
print(df.isnull().sum())

print("\nüîπ Statistical Summary:")
display(df.describe())


if 'Product' in df.columns and 'Sales' in df.columns:
    sales_by_product = df.groupby('Product')['Sales'].sum().reset_index()
    print("\nüí∞ Total Sales by Product:")
    display(sales_by_product)

    # Plot - Sales by Product
    plt.figure(figsize=(8,5))
    sns.barplot(x='Product', y='Sales', data=sales_by_product, palette='viridis')
    plt.title('Total Sales by Product')
    plt.xlabel('Product')
    plt.ylabel('Sales')
    plt.show()

if 'Region' in df.columns and 'Sales' in df.columns:
    sales_by_region = df.groupby('Region')['Sales'].sum().reset_index()
    print("\nüìç Total Sales by Region:")
    display(sales_by_region)

    # Plot - Sales Distribution by Region (Pie Chart)
    plt.figure(figsize=(6,6))
    plt.pie(sales_by_region['Sales'], labels=sales_by_region['Region'], autopct='%1.1f%%', startangle=90)
    plt.title('Sales Distribution by Region')
    plt.show()

print("""
üìà Basic Data Insights:
- Shows total sales per product and per region.
- Visual charts help identify top-performing categories or regions.
- You can modify the column names and groupings as needed.
""")
```
